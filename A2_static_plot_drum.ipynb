# Static Phase Distribution Analysis

This notebook generates static (non-animated) phase distribution plots for drum onset analysis, providing a comprehensive view of rhythmic patterns across different drum types and time windows.

## Overview

The notebook creates static visualizations of drum onset phase distributions, including:
- **Individual plots** for each drum type (Dun, J1, J2)
- **Combined subplots** showing all drum types together
- **Merged/stacked plots** with overlapping distributions

## Features

- **Full duration analysis**: Analyze entire recording duration
- **Time window analysis**: Focus on specific time segments
- **Multiple visualization formats**: Individual, combined, and merged layouts
- **High-resolution output**: Configurable DPI and figure sizes

## Workflow

1. **Configuration**: Set up file paths, time windows, and output directories
2. **Individual Analysis**: Generate separate phase distribution plots for each drum type
3. **Combined Visualization**: Create multi-panel subplots with all drum types
4. **Merged Analysis**: Generate stacked/overlapping distribution plots

## Prerequisites

- Drum onset CSV files (`data/drum_onsets/`)
- Virtual cycle CSV files (`data/virtual_cycles/`)
- Optional: Dance mode pickle files (`data/dance_modes_ts/`) for time segment selection

## Output

All static plots are saved as PNG files to the `output_static_plot/` directory, with filenames indicating the recording name, drum type (if applicable), and time window.
